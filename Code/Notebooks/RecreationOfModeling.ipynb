# Neural Network

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam, SGD, Adadelta
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
def create_eboss_model():
    model = Sequential()
    model.add(Dense(96, input_dim=96, activation='relu'))
    model.add(Dense(20, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))

    model.compile(loss='binary_crossentropy', optimizer=Adam(lr=.001), metrics=['accuracy'])
    return model
Keras_eboss_NN_Model = KerasClassifier(create_eboss_model, epochs=100, batch_size=8, verbose=0)

def create_manga_model():
    model = Sequential()
    model.add(Dense(108, input_dim=108, activation='relu'))
    model.add(Dense(20, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))

    model.compile(loss='binary_crossentropy', optimizer=Adam(lr=.001), metrics=['accuracy'])
    return model
Keras_manga_NN_Model = KerasClassifier(create_manga_model, epochs=100, batch_size=8, verbose=0)
```

```python
Keras_eboss_NN_Model = KerasClassifier(create_eboss_model, epochs=100, batch_size=8, verbose=0)
Keras_manga_NN_Model = KerasClassifier(create_manga_model, epochs=5, batch_size=8, verbose=0)
```

eboss

```python
cvm_eboss_nn = CrossValidationModeler(eboss_data['X'], eboss_data['Y'], Keras_eboss_NN_Model)
cross_val_params = {}
cvm_eboss_nn.run_cross_val(cross_val_params)
cvm_eboss_nn.get_last_run_relevant_metrics().sort_values(by='mean_test_accuracy', ascending=False).head()
```

manga

```python
cvm_manga_nn = CrossValidationModeler(manga_data['X'], manga_data['Y'], Keras_manga_NN_Model)
cross_val_params = {}
cvm_manga_nn.run_cross_val(cross_val_params)
cvm_manga_nn.get_last_run_relevant_metrics().sort_values(by='mean_test_accuracy', ascending=False).head()
```
